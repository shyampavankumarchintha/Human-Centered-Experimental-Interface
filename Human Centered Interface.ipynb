```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    "Time": range(1,101),
    "Temperature": np.random.normal(25,2,100),
    "Humidity": np.random.normal(50,5,100)
})

data.head()
```

```python
import plotly.express as px

fig = px.line(data, x="Time", y="Temperature")
fig.show()
```

```python
data.describe()
```

```python
px.scatter(data,
           x="Temperature",
           y="Humidity")
```

```python
px.line(data,x="Time",y="Temperature")
```

```python
px.line(
    data,
    x="Time",
    y="Temperature",
    markers=True,
    title="Temperature Monitoring Dashboard"
)
```
